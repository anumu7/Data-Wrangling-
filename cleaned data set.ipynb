## replacing some special characters and also changing the dataset type of the score column

```python
df['score'] = df['score'].replace(['9.'], '9.0')
df['score'] = df['score'].replace(['9,.0'], '9.0')
df['score'] = df['score'].replace(['8,9f'], '8.9')
df['score'] = df['score'].replace(['08.9'], '8.9')
df['score'] = df['score'].replace(['8..8'], '8.8')
df['score'] = df['score'].replace(['8:8'], '8.8')
df['score'] = df['score'].replace(['nan'], '0')
df['score'] = df['score'].replace(['++8.7'], '8.7')
df['score'] = df['score'].replace(['8.7.'], '8.7')
df['score'] = df['score'].replace(['8,7e-0'], '8.7')
df['score'] = df['score'].replace(['8,6'], '8.6')
```

## checking, cleaning and replacing some values in the release_year column and changing the type

```python
df['release_year'].tolist()
```

```python
df['release_year'] = df['release_year'].replace(['09 21 1972'], '1972-09-21')
df['release_year'] = df['release_year'].replace([' 23 -07-2008'], '2008-07-23')
df['release_year'] = df['release_year'].replace(['22 Feb 04'], '2004-02-04')
df['release_year'] = df['release_year'].replace(['10-29-99'], '1999-10-29')
df['release_year'] = df['release_year'].replace(['23rd December of 1966 '], '1966-12-23')
df['release_year'] = df['release_year'].replace(['nan'], '0000')
df['release_year'] = df['release_year'].replace(['01/16-03'], '2016-01-03')
df['release_year'] = df['release_year'].replace(['18/11/1976'], '1976-11-18')
df['release_year'] = df['release_year'].replace(['21-11-46'], '1946-11-21')
df['release_year'] = df['release_year'].replace(['1976-13-24'], '1976-01-24')
```

```python
df['release_year'].tolist()
```

```python
df["release_year"] = pd.to_datetime(df["release_year"])
```
